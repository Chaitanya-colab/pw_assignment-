- What are decorators in PYthan? Explain it with an example.Write down its use cases.

````python
"""
Decorators:
    decorators are a powerful tool that allows you to modify the behavior of
    functions without permanently altering their source code. They provide a
    concise and flexible way to add functionality like logging, authentication, or error handling.
"""
import time

def execution_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function '{func.__name__}' executed in {(end_time - start_time):.4f} seconds")
        return result
    return wrapper

@execution_time #<---Decorators
def some_function(a, b):
    time.sleep(1)
    return a * b

result = some_function(2, 3)
"""
Cases of Decorators

Logging:
    Track the execution of functions, recording information like arguments, results, and execution times.
Authentication and Authorization:
    Restrict access to certain functions based on user roles or permissions.
Error Handling:
    Define centralized error handling logic that can be applied to multiple functions.
Caching:
    Store the results of expensive functions to avoid redundant calculations.
Metrics and Monitoring:
    Track and measure the performance of specific functions or code sections.
"""
````

- How does inheritance work in pvthon? Explain all types of inheritance with an example.

````python
"""
Single Inheritance:
        This is the most common and straightforward type.
        A child class inherits from only one parent class.
        It's easy to understand and avoids potential naming conflicts.
"""
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("Generic animal sound")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def make_sound(self):
        print("Woof!")


"""
Multiple Inheritance:
        A child class inherits from multiple parent classes.
        This can be useful for combining functionalities from different classes,
        but it can also lead to naming conflicts and complexity if not used carefully.
"""
class Swimmer:
    def swim(self):
        print("Can swim")

class Flyer:
    def fly(self):
        print("Can fly")

class Duck(Swimmer, Flyer):  # <
    def quack(self):
        print("Quack!")


"""
Multilevel Inheritance:

        A child class inherits from another child class,
        creating a chain of inheritance. This allows for specialization based on
        multiple levels of parent classes.
"""
class Vehicle:
  def __init__(self, make):
    self.make = make

class Car(Vehicle):
  def __init__(self, make, model):
    super().__init__(make)  # <<
    self.model = model


  def charge(self):
    print("Charging electric car...")

"""
Hierarchical Inheritance:
        This involves creating multiple subclasses from a single parent class.
        Each subclass inherits independently from the parent, specializing in its own functionality.
"""
class Shape:
  def __init__(self, color):
    self.color = color

class Square(Shape):
  def __init__(self, color, side_length):
    super().__init__(color)  # <
    self.side_length = side_length

  def get_area(self):
    return self.side_length**2
````

**1.11, Write a program to find prime numbers between 2 to 100 using nested for loops.**

````python
def is_prime(num):
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

prime_numbers = []
for num in range(2, 101):
  if is_prime(num):
    prime_numbers.append(num)

print("Prime numbers between 2 and 100:", prime_numbers)
````

````python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ₹{amount}. Current balance: ₹{self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ₹{amount}. Current balance: ₹{self.balance}")
        else:
            print("Insufficient funds.")

    def inquiry(self):
        print(f"Account Number: {self.account_number}")
        print(f"Current Balance: ₹{self.balance}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0, interest_rate=0.01):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest applied. Current balance: ${self.balance}")


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance=0, overdraft_limit=100):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if self.balance + self.overdraft_limit >= amount:
            self.balance -= amount
            print(f"Withdrew ₹{amount}. Current balance: ₹{self.balance}")
        else:
            print("Exceeds overdraft limit.")


# Testing
if __name__ == "__main__":
    savings_acc = SavingsAccount("PI2348234098", 1000, 0.05)
    checking_acc = CheckingAccount("CHK-001", 500, 200)

    savings_acc.deposit(500)
    savings_acc.apply_interest()


    checking_acc.deposit(200)
    checking_acc.withdraw(700)
    checking_acc.inquiry()
````

````python
from collections import Counter
import string

def count_word_occurrences(filename):
    try:
        with open(filename, 'r') as file:

            text = file.read().lower()

            text = text.translate(str.maketrans('', '', string.punctuation))

            words = text.split()

            word_counts = Counter(words)


            print("Word Counts:")
            for word, count in word_counts.items():
                print(f"{word}: {count}")

    except FileNotFoundError:
        print(f"The file {filename} does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")


filename = "paragraph.txt"

count_word_occurrences(filename)
````

49.Make a fully functional web application using flask, Mangodb. Signup,Signin page.And after successfully
login .Say hello Geeks message at webpage.

````python
from flask import Flask, render_template, request, redirect, url_for, session
from flask_pymongo import PyMongo
from werkzeug.security import generate_password_hash, check_password_hash

app = Flask(__name__)
app.secret_key = "supersecretkey"

app.config["MONGO_URI"] = "mongodb://localhost:27017/flask_mongo_app"

mongo = PyMongo(app)

@app.route('/')
def home():
    if 'username' in session:
        return f'Hello, {session["username"]}! Welcome to Geeks.'
    return redirect(url_for('login'))

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        user = mongo.db.users.find_one({'username': username})

        if user and check_password_hash(user['password'], password):
            session['username'] = username
            return redirect(url_for('home'))
        return 'Invalid username or password'

    return render_template('login.html')

@app.route('/signup', methods=['GET', 'POST'])
def signup():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        hash_pass = generate_password_hash(password)

        mongo.db.users.insert_one({'username': username, 'password': hash_pass})
        session['username'] = username
        return redirect(url_for('home'))

    return render_template('signup.html')

@app.route('/logout')
def logout():
    session.pop('username', None)
    return redirect(url_for('login'))

if __name__ == '__main__':
    app.run(debug=True)
````

````python
"""
L1 regularization (Lasso Regression): 
    Penalizes the sum of the absolute values of the model's coefficients (weights). 
    L1: Shrinks coefficients towards zero, and can even drive some coefficients to exactly zero. 
    L1 is more robust to outliers in the data because large deviations have less impact on the absolute value penalty.
    
L2 regularization (Ridge Regression): 
    Penalizes the sum of the squared values of the coefficients
    L2: Shrinks coefficients towards zero, but typically not all the way to zero
    L2 is generally computationally faster to optimize than L1.
"""
````

What is a confusion matrix and how is it used?

````python
"""
Confusion:
    A confusion matrix is a visualization tool used in machine learning to evaluate the performance of a classification model. 
    It's particularly helpful for understanding how well the model is classifying data points into different categories.

How is it used:

                 |  Predicted Positive	   |   Predicted Negative          
Actual Positive	 |   True Positive (TP)	   |    False Negative (FN)
Actual Negative	 |   False Positive (FP)   |    True Negative (TN)

True Positives (TP): These are data points where the model correctly predicted the positive class.
True Negatives (TN): These are data points where the model correctly predicted the negative class.
False Positives (FP): These are data points where the model incorrectly predicted the positive class (also known as Type I error).
False Negatives (FN): These are data points where the model incorrectly predicted the negative class (also known as Type II error).
"""
````

Describe the process of constructing a decision tree.

````python
"""
Define the Problem: 
        we need to clearly identify the decision you're trying to make. What are you trying to predict or classify? This forms the root node of your tree.

Choose the Best Splitting Feature:
        At each step, you'll want to identify the most important feature in your data that best separates the data points into distinct groups. This is often done using metrics like Gini impurity or entropy, which measure how well a specific feature separates the data into the desired categories.

Split the Data:
        Based on the chosen feature, you'll split your data into subsets. Each branch will represent a different possibility for the feature's value.

Repeat and Refine:
        For each new subset of data, you'll repeat steps 2 and 3. You'll keep asking questions and splitting the data further until you reach a point where
"""
````

Describe the working principle of a decision tree.

````python
"""
Structure:

        Decision Nodes:
                These contain questions or tests based on features of your data. They act like branching points in the flowchart.
                
        Leaf Nodes: 
                These represent the final outcomes or classifications you're aiming for. They don't have any further branches.

Making Predictions: 
        When you use the tree for prediction, you start at the root node (the topmost question). 

Following the Path: 
        At each decision node, you answer the question based on the data point's features. 

The Answer Lies in the Leaves: 
        The final outcome or prediction is determined by the label of the leaf node you reach.
"""
````

What is information gain and how is it used in decision trees?

````python
"""
Entropy: 
        Information gain is closely linked to entropy, which is a measure of impurity in a dataset. High entropy indicates a dataset with mixed classifications (like having a mix of apples and oranges in a basket).

Information Gain Calculation: 
        To decide how well a feature separates the data, we calculate the information gain it brings. This involves:

Entropy before Split: 
        We calculate the overall entropy of the data at the current node (like the mix of fruits in the basket before separating).
        
Entropy after Split: 
        We then consider the entropy of each subset created by splitting on the chosen feature (like the entropy of the apple pile and the orange pile).
        
Gain in Purity: 
        Information gain is the difference between the entropy before the split (higher impurity) and the weighted average of the entropy after the split (considering the size of each subset). 

Choosing the Best Split:
        When building the tree, we calculate information gain for all possible features at each node. 
"""
````

Explain Gini impurity and its role in decision trees.

````python
"""
What it Represents:
        Gini impurity calculates the probability of a randomly chosen data point being misclassified if labelled based on the proportion of classes present in the node.
        It ranges from 0 to 0.5.
        0: Perfectly pure node (all data points belong to the same class).
        0.5: Maximum impurity (equal distribution of classes).
        
Role in Decision Tree Building:
        During tree construction, Gini impurity is calculated for all possible splits on each feature at a node.
        The feature that leads to the lowest Gini impurity after the split is chosen. 
        By prioritizing features that minimize Gini impurity, the decision tree progressively separates the data into more homogenous groups, 
"""
````

````python
"""
Reduced Overfitting: 
    Decision trees are prone to overfitting, meaning they can become too focused on the specific training data and perform poorly on unseen data.

        Each tree is trained on a random subset of the data (with replacement, so some data points appear multiple times).
        The trees also consider only a random subset of features at each split point.
        
        
Improved Accuracy:  
        combining multiple decision trees, a random forest leverages the strengths of each individual tree while reducing their weaknesses. 
        This often leads to higher prediction accuracy compared to a single decision tree.

Handles Missing Data: 
        Random forests can handle missing data points more gracefully than decision trees. 
        Each tree can make a prediction by considering only the available features for a particular data point.
"""
````

Describe the process of gradient boosting in XGBoost.

````python
"""
Build a base model: 
        Predict the observations in the training dataset using an initial model, F0
        
Calculate pseudo residuals:
        Subtract the predicted value from the observed value to calculate the pseudo residuals
        
Build a model on the residuals:
        Fit a new model, h1, to the pseudo residuals from the previous step
        
Combine the models: 
        Combine F0 and h1 to create F1, the boosted version of F0
        
Update predictions: 
        Use the output from the new tree to improve the final model output
"""
````
